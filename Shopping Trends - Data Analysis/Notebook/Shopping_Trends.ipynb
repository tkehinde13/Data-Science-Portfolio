# Credit Card Approval Analysis
### A Data Analysis and Machine Learning Project
#### This project analyses credit card approval data to uncover insights and build predictive models for classification.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Data Loading and Initial Exploration

```python
# Load the shopping trends dataset
data = pd.read_csv("/shopping_trends.csv")
```

```python
# Display the first five rows of the dataset
data.head()
```

```python
# Display dataset information (column names, data types, non-null counts)
data.info()
```

```python
# Generate summary statistics for numerical columns
data.describe()
```

```python
# Display the number of rows and columns in the dataset
data.shape
```

```python
# Count missing (NA) values for each column
data.isna().sum()
```

```python
# Check for duplicated rows in the dataset
data.duplicated().sum()
```

### Exploratory Data Analysis

```python
# Plot a histogram to visualize the distribution of Age
data["Age"].plot(kind="hist")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Bar chart to visualize gender distribution
data["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Group data by Category and calculate the average purchase amount, then visualize it using a bar chart
data.groupby("Category")["Purchase Amount (USD)"].mean().plot(kind="bar")
plt.title("Average Purchase Amount by Category")
plt.xlabel("Category")
plt.ylabel("Average Purchase Amount")
plt.show()
```

```python
# Sort the dataset by Review Rating in descending order
data.sort_values(by="Review Rating", ascending=False, inplace=True)
```

```python
# Pie chart to show the distribution of Subscription Status
data["Subscription Status"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Subscription Status Distribution")
plt.show()
```

```python
# Pie chart displaying the distribution of the top 5 most frequent colors
data["Color"].value_counts().head().plot(kind="pie", autopct="%1.1f%%")
plt.title("Color Distribution")
plt.show()
```

```python
# Compute and display the average purchase amount per season
data.groupby("Season")["Purchase Amount (USD)"].mean()
```

```python
# Compute and display the total purchase amount per season
data.groupby("Season")["Purchase Amount (USD)"].sum()
```

```python
# Bar chart showing the average purchase amount for each season
data.groupby("Season")["Purchase Amount (USD)"].mean().plot(kind="bar")
plt.title("Average Purchase Amount by Season")
plt.xlabel("Season")
plt.ylabel("Average Purchase Amount")
plt.show()
```

```python
# Boxplot to visualize purchase amount distribution based on purchase frequency
data.boxplot(column="Purchase Amount (USD)", by="Frequency of Purchases")
plt.title("Purchase Amount by Frequency of Purchases")
plt.xlabel("Frequency of Purchases")
plt.ylabel("Purchase Amount")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Bar chart displaying the most preferred payment methods
data["Payment Method"].value_counts().plot(kind="bar")
plt.title("Preferred Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

### Data Aggregation

```python
# Create a pivot table showing the average purchase amount per item in different locations
pd.pivot_table(data, values="Purchase Amount (USD)", index="Location", columns="Item Purchased", aggfunc="mean").head()
```

```python
# Compute total purchase amount for each payment method
data.groupby("Payment Method")["Purchase Amount (USD)"].sum()
```

```python
# Compute average purchase amount for each size category
data.groupby("Size")["Purchase Amount (USD)"].mean()
```

```python
# Compute the average review rating when a discount is applied
data.groupby("Discount Applied")["Review Rating"].mean()
```

```python
# Count the number of times each color appears in different categories
data.groupby("Color")["Category"].value_counts()
```

```python
# Compute the median of previous purchases grouped by frequency of purchases
data.groupby("Frequency of Purchases")["Previous Purchases"].median()
```

```python
# Compute the average review rating for each season
data.groupby("Season")["Review Rating"].mean()
```

```python
# Bar chart showing the count of purchases for each season
data["Season"].value_counts().plot(kind="bar")
plt.ylabel("Count")
plt.xlabel("Season")
plt.title("Count of Purchases for each Season")
plt.show()
```

```python
# Pie chart displaying the count of purchases for each category
data["Category"].value_counts().plot(kind="pie")
plt.ylabel("Count")
plt.xlabel("Category")
plt.title("Count of Purchases for each Category")
plt.show()
```

### Advanced Visualisations

```python
# Boxplot for review rating distribution by gender
data.boxplot(column="Review Rating", by="Gender")
plt.title("Review Rating Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Review Rating")
plt.show()
```

```python
# Histogram to visualize the distribution of purchase amounts
data["Purchase Amount (USD)"].plot(kind="hist", bins=10)
plt.title("Purchase Amount Distribution")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart showing the average review rating per color
data.groupby("Color")["Review Rating"].mean().head().plot(kind="bar")
plt.title("Review Rating by Color")
plt.xlabel("Color")
plt.ylabel("Review Rating")
plt.show()
```

```python
# Pie chart showing the percentage of total purchase amount per payment method
data.groupby("Payment Method")["Purchase Amount (USD)"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Purchase Amount by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("")
plt.show()
```

```python
# Boxplot showing purchase amount distribution by frequency of purchases
data.boxplot(column="Purchase Amount (USD)", by="Frequency of Purchases")
plt.title("Purchase Amount Distribution by Frequency of Purchases")
plt.xlabel("Frequency of Purchases")
plt.ylabel("Purchase Amount")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Pie chart displaying the total purchase amount percentages for each season
data.groupby("Season")["Purchase Amount (USD)"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Purchase Amount by Season")
plt.xlabel("Season")
plt.ylabel("")
plt.show()
```
